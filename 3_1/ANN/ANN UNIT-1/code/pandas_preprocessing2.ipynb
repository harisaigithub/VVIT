```python
#single feature scaling
df['age']=df['age']/df['age'].max()
```

```python
df.head()
```

```python
#MinMax
df['water']=(df['water']-df['age'].min())/(df['water'].max()-df['age'].min())
```

```python
#z-score
df['slag']=(df['slag']-df['slag'].mean())/df['slag'].std()
```

```python
import numpy as np
df['cement']=np.log(df['cement'])
```

```python
from scipy.stats import zscore
df['fineagg']=zscore(df['fineagg'])
```
